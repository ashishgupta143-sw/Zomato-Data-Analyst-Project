```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of ratings
plt.figure(figsize=(10, 6))
sns.countplot(x='rate', data=df, palette='Blues')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Count plot for online orders and table bookings
plt.figure(figsize=(10, 6))
sns.countplot(x='online_order', data=df, palette='Set1', label='Online Orders')
sns.countplot(x='book_table', data=df, palette='Set2', label='Table Bookings', alpha=0.5)
plt.title('Online Orders vs Table Bookings')
plt.xlabel('Option')
plt.ylabel('Number of Restaurants')
plt.legend()
plt.show()
```

```python
# Boxplot for votes by restaurant type
plt.figure(figsize=(12, 6))
sns.boxplot(x='rest_type', y='votes', data=df, palette='viridis')
plt.title('Votes by Restaurant Type')
plt.xlabel('Restaurant Type')
plt.ylabel('Number of Votes')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert approximate cost to numeric (remove '' and 'string' part)
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].replace('[\,]', '', regex=True).astype(float)

# Boxplot for approximate cost by cuisine type
plt.figure(figsize=(12, 6))
sns.boxplot(x='cuisines', y='approx_cost(for two people)', data=df, palette='coolwarm')
plt.title('Approximate Cost for Two People by Cuisine Type')
plt.xlabel('Cuisine Type')
plt.ylabel('Approximate Cost ()')
plt.xticks(rotation=45)
plt.show()
```
